```python
#импортируем нужные пакеты
import numpy as np
import pandas as pd

#создает датафрейм из локальной версии файла
df = pd.read_csv('HR.csv')

"""
Описание колонок:

1. satisfaction_level - Уровень удовлетворенности работой
2. Last_evaluation - Время с момента последней оценки в годах
3. number_projects - Количество проектов, выполненных за время работы
4. average_monthly_hours - Среднее количество часов на рабочем месте в месяц
5. time_spend_company - Стаж работы в компании в годах
6. work_accident - Происходили ли несчастные случаи на рабочем месте с сотрудником
7. left - уволился ли сотрудник
8. promotion_last_5years - повышался ли сотрудник за последние пять лет
9. department - отдел в котором работает сотрудник
10. salary - относительный уровень зарплаты

"""
df.info()
```

```python
#выделим столбцы с категориальными данными
w_accident = df['Work_accident']
left = df['left']
promo_l_5 = df['promotion_last_5years']
department = df['department']
salary = df['salary']
```

```python
#выделим столбцы с количественными данными
satisfaction = df['satisfaction_level']
last_evaluation = df['last_evaluation']
number_project = df['number_project']
average_montly_hours = df['average_montly_hours']
time_spend_company = df['time_spend_company']
```

Рассчитайте основные статистики для переменных
(среднее,медиана,мода,мин/макс,сред.отклонение)

```python
#для столбца satisfaction_level рассчитаем базовые статистики: среднее, медиану, моду, максимальную и минимальную величины, среднее отклонение
s_l_mean = satisfaction.mean()
s_l_median = satisfaction.median()
s_l_mode = satisfaction.mode()[0]
s_l_min = satisfaction.min()
s_l_max = satisfaction.max()
s_l_std = satisfaction.std()

#выведем результат
print('Расчет базовых статистик по столбцу satisfaction_level:')
print(f'Среднеарифметическое: {s_l_mean}')
print(f'Медиана: {s_l_median}')
print(f'Мода: {s_l_mode}')
print(f'Минимальное значения: {s_l_min}')
print(f'Максимальное значения: {s_l_max}')
print(f'Среднее отклонение: {s_l_std}')
```

```python
#для столбца last_evaluation рассчитаем базовые статистики: среднее, медиану, моду, максимальную и минимальную величины, среднее отклонение
l_e_mean = last_evaluation.mean()
l_e_median = last_evaluation.median()
l_e_mode = last_evaluation.mode()[0]
l_e_min = last_evaluation.min()
l_e_max = last_evaluation.max()
l_e_std = last_evaluation.std()

#выведем результат
print('Расчет базовых статистик по столбцу last_evaluation:')
print(f'Среднеарифметическое: {l_e_mean}')
print(f'Медиана: {l_e_median}')
print(f'Мода: {l_e_mode}')
print(f'Минимальное значения: {l_e_min}')
print(f'Максимальное значения: {l_e_max}')
print(f'Среднее отклонение: {l_e_std}')
```

Рассчитайте сколько сотрудников работает в каждом
департаменте.

```python
#посчитаем количество сотрудников сгруппировав по департаменту; для удобства оставим только 1 столбец, например, зарплаты
df_dep = df.groupby('department').count()[['salary']]

#создадим новый series и переименум столбец на корректный
employees = df_dep.copy()
employees = employees.rename(columns={'salary': 'number_of_the_employees'})

#выведем результат по количеству сотрудников в департаменте
employees
```

Показать распределение сотрудников по зарплатам.

```python
#посчитаем распределение зарплат по всем сотрудникам
df_sal = df.groupby('salary').count()[['left']]
employ = df_sal.copy()
employ = employ.rename(columns={'left': 'employees'})

#распределение зарплат по всем сотрудникам
employ
```

```python
#импортируем нужные пакеты для визуализации
import matplotlib.pyplot as plt
from matplotlib import colorbar

employ.sort_values(by='salary', ascending=True).plot(kind='bar',
                                                                  title='Распределение сотрудников по зарплатам',
                                                                  xlabel='salary',
                                                                  ylabel='employees', grid =True)
plt.show()
```
